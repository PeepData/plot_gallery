### Import libraries 

```python
import numpy as np
import pandas as pd

from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, NumeralTickFormatter, Title, Text, LabelSet, FactorRange, SingleIntervalTicker

output_notebook()
```

#### Colors 

```python
GRAY1, GRAY2, GRAY3 = '#231F20', '#414040', '#555655'
GRAY4, GRAY5, GRAY6 = '#646369', '#76787B', '#828282'
GRAY7, GRAY8, GRAY9 = '#929497', '#A6A6A5', '#BFBEBE'
BLUE1, BLUE2, BLUE3, BLUE4 = '#174A7E', '#4A81BF', '#94B2D7', '#94AFC5'
RED1, RED2, RED3 = '#800000', '#C3514E', '#E6BAB7', 
GREEN1, GREEN2 = '#0C8040', '#9ABB59'
ORANGE1 = '#F79747'
```

#### Font 

```python
FONT = 'Arial'
```

#### Preprocessing 

```python
dataset = pd.read_csv('../data/data_fig_0601.csv')
dataset.shape
```

```python
dataset.head()
```

#### Plot 

```python
# Set the source of the plot
source = ColumnDataSource(dataset)


# Create the figure
p = figure(x_range=(0, 40), 
           y_range=(0, 65000), 
           plot_height=550, 
           plot_width=930, 
           title='Annual giving campaing promise',
           toolbar_location='above')


# Add last year line
p.line(x='Days since launch', 
       y='Last year', 
       line_color=BLUE3,
       line_width=1.5,
       source=source)

p.circle(x=30, 
         y=51400, 
         color=BLUE3, 
         size=7)

# Add progress to date line
p.line(x='Days since launch', 
       y='Progress to date', 
       line_color=BLUE2,
       line_width=4,
       source=source)

p.circle(x=10, 
         y=33967, 
         color=BLUE2, 
         size=13)


# Text annotations
# This is a weakness of bokeh, you cannot format a single word within a text. We have to do it separately.

# Last year annotation
p.text(x=30, 
       y=50000,
       x_offset=0,
       y_offset=-2,
       text=['Last year\n \t$51,400'],
       text_color=BLUE2,
       text_font=FONT,
       text_font_size='13pt')

# Progress to date annotation
p.text(x=6, 
       y=35000,
       x_offset=-4,
       y_offset=-10,
       text=['Progress to date'],
       text_color=BLUE2,
       text_font=FONT,
       text_font_size='18pt')

p.text(x=10, 
       y=35000,
       x_offset=15,
       y_offset=20,
       text=['$33,967'],
       text_color=BLUE2,
       text_font=FONT,
       text_font_size='16pt',
       text_font_style='bold')

# Goal annotation
p.text(x=5, 
       y=50000,
       x_offset=-35,
       y_offset=10,
       text=['GOAL'],
       text_color=GRAY2,
       text_font=FONT,
       text_font_size='12pt')

p.segment(x0=0, 
          y0=50000, 
          x1=3,
          y1=50000, 
          color=GRAY3, 
          line_width=1.5)

p.segment(x0=6, 
          y0=50000, 
          x1=32,
          y1=50000, 
          color=GRAY3, 
          line_width=1.5)


# Elements attributes

# Modify title attributes
p.title.offset = -80
p.title.text_color = GRAY4
p.title.text_font = FONT
p.title.text_font_size = '20pt'
p.title.text_font_style = 'normal'


# Modify x axis attributes
p.xaxis.bounds = (0, 30)
p.xaxis.axis_line_color = GRAY9
p.xaxis.axis_label = 'Days since campaign launch' + ' ' * 142
p.xaxis.axis_label_standoff = 5  
p.xaxis.axis_label_text_color = GRAY6
p.xaxis.axis_label_text_font = FONT
p.xaxis.axis_label_text_font_size = '12pt'
p.xaxis.axis_label_text_font_style = 'normal'
p.xaxis.major_label_standoff = 5
p.xaxis.major_label_text_color = GRAY6
p.xaxis.major_label_text_font = FONT
p.xaxis.major_label_text_font_size = '11pt'
p.xaxis.major_tick_in = 0
p.xaxis.major_tick_out = 3
p.xaxis.major_tick_line_color = GRAY9
p.xaxis.minor_tick_line_color = None
p.xaxis.ticker = SingleIntervalTicker(interval=5, num_minor_ticks=6)
p.xgrid.grid_line_color = None


# Modify y axis attributes
p.yaxis.bounds = (0, 60000)
p.yaxis.formatter = NumeralTickFormatter(format='%$0,0%')
p.yaxis.axis_line_color = GRAY9
p.yaxis.axis_label = ' ' * 67 + 'Money raised' 
p.yaxis.axis_label_standoff = 5  
p.yaxis.axis_label_text_color = GRAY6
p.yaxis.axis_label_text_font = FONT
p.yaxis.axis_label_text_font_size = '12pt'
p.yaxis.axis_label_text_font_style = 'normal' 
p.yaxis.major_label_standoff = 5
p.yaxis.major_label_text_color = GRAY7
p.yaxis.major_label_text_font = FONT
p.yaxis.major_label_text_font_size = '11pt'
p.yaxis.major_tick_in = 0
p.yaxis.major_tick_out = 3
p.yaxis.major_tick_line_color = GRAY9
p.yaxis.minor_tick_line_color = None
p.ygrid.grid_line_color = None


# Handle backgrounds color
p.background_fill_color = 'white'
p.border_fill_color = 'white'
p.outline_line_color = 'white'


show(p)
```
