```python
from bokeh.io import output_notebook, show
from bokeh.layouts import column
from bokeh.plotting import figure

output_notebook()

x = list(range(11))
y0 = x
y1 = [10 - i for i in x]
y2 = [abs(i - 5) for i in x]

# create three plots
s1 = figure(plot_width=800, plot_height=200, background_fill_color="#fafafa")
s1.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=2)

s2 = figure(plot_width=800, plot_height=200, background_fill_color="#fafafa")
s2.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=2)

s3 = figure(plot_width=800, plot_height=200, background_fill_color="#fafafa")
s3.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=2)

# put the results in a row and show
show(column(s1, s2, s3))
```

```python
from bokeh.models import FactorRange
from bokeh.plotting import figure

factors = [
    ("Q1", "jan"), ("Q1", "feb"), ("Q1", "mar"),
    ("Q2", "apr"), ("Q2", "may"), ("Q2", "jun"),
    ("Q3", "jul"), ("Q3", "aug"), ("Q3", "sep"),
    ("Q4", "oct"), ("Q4", "nov"), ("Q4", "dec"),

]

p = figure(x_range=FactorRange(*factors), plot_height=350,
           toolbar_location=None, tools="")

x = [ 10, 12, 16, 9, 10, 8, 12, 13, 14, 14, 12, 16 ]
p.vbar(x=factors, top=x, width=0.9, alpha=0.5, legend_label='hola')

p.line(x=["Q1", "Q2", "Q3", "Q4"], y=[12, 9, 13, 14], color="red", line_width=2)

p.below[0].group_text_font_size = '16px'
p.below[0].group_text_font_style = 'normal'
p.below[0].group_text_color = 'green'
p.below[0].major_label_text_font_size = '14px'
p.below[0].major_label_text_color = 'red'
p.xaxis[0].separator_line_color = None
p.x_range.range_padding = 0.1
p.legend.location = 'top_left'
show(p)
```

```python
p = figure(x_range=(1, 5), y_range=(-5, 7))

# Note that ymin and ymax are in reverse order in y_range.
p.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=2)

p.multi_polygons(xs=[[[[4, 4, 5, 5]]]],
                 ys=[[[[0, -1, -1, 0]]]],                 
                 fill_alpha=0.2,
                 fill_color='blue',
                 line_color=None)

p.xaxis.fixed_location = 0
p.yaxis.bounds = (0, 7)
show(p) 
```

```python
from bokeh.io import output_file, show
from bokeh.plotting import figure

fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries']
years = ["2015", "2016", "2017"]
colors = ["#c9d9d3", "#718dbf", "#e84d60"]

data = {'fruits' : fruits,
        '2015'   : [2, 1, 4, 3, 2, 4],
        '2016'   : [5, 3, 4, 2, 4, 6],
        '2017'   : [3, 2, 4, 4, 5, 3]}

p = figure(x_range=fruits, plot_height=250, title="Fruit Counts by Year",
           toolbar_location=None, tools="hover", tooltips="$name @fruits: @$name")

r = p.vbar_stack(years, x='fruits', width=0.9, color=colors, source=data,
             legend_label=years)

p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xgrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.legend.location = "top_left"
p.legend[0].orientation = "horizontal"

p.legend.items[1].label = None

show(p)
```
