#### Preprocessing 

```python
dataset = pd.read_csv('../../data/data_figure_6.csv')
dataset.shape
```

```python
dataset.head()
```

```python
dataset_hbar = dataset[['Segment', 'US Population', 'Our Customers']].set_index('Segment').T.rename_axis('Customers').sort_values('Customers', ascending=True).reset_index()
```

```python
dataset_hbar.columns.name = None
```

```python
dataset_hbar
```

#### Colors 

```python
GRAY_1, GRAY_2, GRAY_3 = '#A6A6A5', '#929497', '#231F20'
GRAY_4, GRAY_5, GRAY_6 = '#838383', '#555655', '#828282'
GRAY_7, GRAY_8, GRAY_9 = '#646369', '#333333', '#9c9c9c'
GRAY_10 = '#76787B'
BLUE_1, BLUE_2, BLUE_3 = '#94B2D7', '#4A81BF', '#174A7E'
ORANGE_1 = '#F79747'
```

#### Font 

```python
FONT = 'Arial'
```

#### Plot 

```python
# Set the source of the plot
source = ColumnDataSource(dataset_hbar)

colors = [GRAY_9, GRAY_9, BLUE_2, BLUE_2, BLUE_2, GRAY_9, GRAY_9]

segment = dataset['Segment'].to_list()
customers = dataset_hbar['Customers'].to_list()


# Set color by category
#colors = [GRAY_4, GRAY_4, ORANGE_1, BLUE_2, BLUE_2]


# Create the figure
p = figure(#x_range=(-17, 105),
           #x_axis_location='above',
           x_range=customers,
           plot_height=500,
           plot_width=650, 
           title='Pilot program was a success' + ' ' * 45,
           toolbar_location='below')


# Add bars to the figure
p.vbar_stack(segment, 
             x='Customers',
             color=colors,
             width=0.7, 
             line_width=1.2, 
             line_color='white',
             source=source)




show(p)
```
